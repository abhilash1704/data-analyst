##  Seaborn is a data visualization library in Python built on top of Matplotlib.
## It provides a high-level interface for creating attractive, informative, and statistical plots with much less code compared to plain Matplotlib.

```python
# distrution plots

import seaborn as sns 
import matplotlib.pyplot as plt
sns.histplot(data=[1,2,2,3,3,3,4,4,5], kde=True)
plt.show()
```

```python
import seaborn as sns
import seaborn as sns

sns.set(style="ticks")

# Loading the dataset
df = sns.load_dataset("anscombe")

# Show the results of a linear regression
sns.lmplot(x="x", y="y", data=df)
```

```python
import seaborn as sns


sns.set(style="dark")
fmri = sns.load_dataset("fmri")

# Plot the responses for different\
# events and regions
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             style="event",
             data=fmri)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

data = sns.load_dataset("iris") 

plt.figure(figsize = (8, 6))

sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

sns.despine()

plt.show()
```
